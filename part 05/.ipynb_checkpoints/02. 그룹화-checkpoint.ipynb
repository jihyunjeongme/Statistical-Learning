# 그룹화 

## k-means 클러스터링

### 데이터 셋(Dataset) 및 전처리

```python
# 연산 처리를 위한 패키지
import pandas as pd
import numpy as np

# 그룹화를 위한 패키지
from sklearn.cluster import KMeans

# 시각화를 위한 패키지
from matplotlib import pyplot as plt
import seaborn as sns

# 데이터 셋
from sklearn import datasets

# 그래프를 실제로 그리기 위한 설정
%matplotlib inline

# 경고 메세지 무시
import warnings
warnings.filterwarnings('ignore')

# iris dataset을 불러옵니다.
iris = datasets.load_iris()
df_iris = pd.DataFrame(data=np.c_[iris.data, iris.target], columns= iris.feature_names + ['target'])

# target을 분류
df_iris['target'] = df_iris['target'].apply(lambda x: 'setosa' if 0 == x else ('versicolor' if 1 == x else 'virginica'))
df_iris.head()
```

```python
sns.pairplot(data=df_iris.loc[:, ['target', 'sepal length (cm)','sepal width (cm)']], hue="target", size=5,
             plot_kws=dict(s=50, linewidth=1))
```

### k를 3으로 군집 개수 지정

```python
# k를 3개로 
cluster = KMeans(n_clusters = 3, n_jobs = -1, random_state=0)

# 2차원 평면상에서 시각화를 위해 2개의 속성만 사용
model = cluster.fit(df_iris[['sepal length (cm)','sepal width (cm)']])

# 각각의 클러스터의 중심 확인
centers = model.cluster_centers_
print(centers)
```

```python
model.labels_
```

```python
df_iris['km_cluster'] = model.labels_
df_iris['km_cluster'] = df_iris['km_cluster'].apply(lambda x: 'setosa' if 0 == x else ('versicolor' if 1 == x else 'virginica'))

df_iris.head()
```

```python
pd.DataFrame(df_iris.groupby(['target','km_cluster'])[['sepal length (cm)','sepal width (cm)']].count())
```

```python
sns.pairplot(data=df_iris.loc[:, ['km_cluster', 'sepal length (cm)','sepal width (cm)']], hue="km_cluster", size=5,
             plot_kws=dict(s=50, linewidth=1))
```

### 군집화를 위한 가상 데이터 생성


- 인수:
    - n_samples : 표본 데이터의 수
    - n_features : 독립 변수의 수
    - centers : 생성할 클러스터의 수 혹은 중심
    - cluster_std: 클러스터의 표준 편차

- 반환값:
    - X : [n_samples, n_features] 크기의 배열
    - 독립 변수
    - y : [n_samples] 크기의 배열
    - 종속 변수

#### k를 5로 할 경우

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=5, random_state=27)

df_k5 = pd.DataFrame(data=X, columns=['x1','x2'])
df_k5['target'] = y
df_k5.head()
```

```python
plt.scatter(df_k5['x1'], df_k5['x2'], marker='o', c=y, s=50,
            edgecolor="k")
plt.show()
```

```python
# k-means 클러스터링 
cluster = KMeans(n_clusters = 5, n_jobs = -1, max_iter=200, random_state=27)
pred = cluster.fit_predict(X)
df_k5['pred']= pred

pd.DataFrame(df_k5.groupby(['target','pred'])[['x1','x2']].count())
```

```python
# 그룹화된 클러스터 1 ~ 5
plt.scatter(
    X[pred == 0, 0], X[pred == 0, 1],
    s=50, c='green',
    marker='o', label='cluster 1'
)

plt.scatter(
    X[pred == 1, 0], X[pred == 1, 1],
    s=50, c='orange',
    marker='s', label='cluster 2'
)

plt.scatter(
    X[pred == 2, 0], X[pred == 2, 1],
    s=50, c='gold',
    marker='v', label='cluster 3'
)

plt.scatter(
    X[pred == 3, 0], X[pred == 3, 1],
    s=50, c='brown',
    marker='h', label='cluster 4'
)

plt.scatter(
    X[pred == 4, 0], X[pred == 4, 1],
    s=50, c='violet',
    marker='d', label='cluster 5'
)

# 클러스터 별 중심 위치
plt.scatter(
    cluster.cluster_centers_[:, 0], cluster.cluster_centers_[:, 1],
    s=100, c='r',
    marker='*', label='centroids'
)

plt.legend()
plt.grid()
plt.show()
```

```python
inertia = list()

# inertia를 측정
for i in range(1,11):
    model = KMeans(n_clusters=i)
    model.fit(X)
    inertia.append(model.inertia_)

# K수에 따라 inertia의 변화
plt.plot(range(1,11), inertia, '-o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.show()
```

## 계층적 군집 분석

### 군집화를 위한 가상 데이터 생성

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=14, n_features=2, centers=2, random_state=6, cluster_std=0.3)

df_hc = pd.DataFrame(data=X, columns=['x1','x2'])
df_hc['target'] = y
df_hc.head()
```

```python
plt.scatter(df_hc['x1'], df_hc['x2'], marker='o', c=y, s=20)

labeling = range(1,15)
for label, x, y in zip(labeling, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x,y), xytext=(-2,2),
        textcoords='offset points', ha='left', va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# 단일 연결법: 주어진 군집의 관측치들의 거리를 계산
linked = linkage(X, 'single')
labels = range(1, 15)

plt.figure(figsize=(10,6))
dendrogram(linked,
           orientation='top',
           labels=labels)
plt.show()
```

```python
# 와드 방법: 각 군집이 합쳐졌을 때 군집중심(군집 내 관측치들의 평균)과 각 관측치 값의 차이 제곱을 거리로 이용한다. 
linked = linkage(X, 'ward')
labels = range(1, 15)

plt.figure(figsize=(6,4))
dendrogram(linked,
           orientation='right',
           labels=labels)
plt.show()
```

### 예제 - House Sales in King County, USA

```python
import scipy.cluster.hierarchy as shc

df = pd.read_csv('./data/regression/kc_house_data.csv')
df.head()
```

### date 데이터 처리 및 새로운 변수 추가

```python
df['date2'] = df['date'].apply(lambda x: x[0:4])
df['date2'] = df['date2'].astype('int64')
df['date2']

df['sold-built_years'] = df.apply(lambda x: ((x['date2']) - (x['yr_built'])), axis=1)
df['sold-built_years']
```

```python
X = df[['bedrooms','sqft_living','waterfront','view','sold-built_years']]
y = df['price']
X
```

```python
# 500개만 선택
data = X[:500]

# Ward's method: 군집 중심(군집 내 관측치들의 평균)과 관측치 값의 차이 제곱을 거리로 이용
plt.figure(figsize=(12,10))
dendrogram = shc.dendrogram(shc.linkage(data, method='ward'), orientation='right')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
model = cluster.fit_predict(data)

print("Cluster 1의 평균 가격(price): ", np.mean([x for x, y in zip(y, cluster.fit_predict(data)) if y == 0]))
print("Cluster 2의 평균 가격(price): ", np.mean([x for x, y in zip(y, cluster.fit_predict(data)) if y == 1]))    
```

```python
data.iloc[cluster.fit_predict(data) == 0, :].describe()
```

```python
data.iloc[cluster.fit_predict(data) == 1, :].describe()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(
    data[model == 0]['sqft_living'], data[model == 0]['sold-built_years'],
    s=40, c='green',
    marker='o', label='cluster 1'
)

plt.scatter(
    data[model == 1]['sqft_living'], data[model == 1]['sold-built_years'],
    s=40, c='gold',
    marker='s', label='cluster 2'
)

plt.xlabel('sqft_living')
plt.ylabel('sold-built_years')

plt.legend()
plt.grid()
plt.show()
```

```python
df_hc = data
df_hc['clustering'] = model

sns.pairplot(df_hc, hue="clustering", size=3,
             plot_kws=dict(s=50, linewidth=1))
```
